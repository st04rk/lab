```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread("gtr.jpeg")

blur_image=cv2.GaussianBlur(img,(3,3),0)
edges=cv2.Canny(image=blur_image,threshold1=100,threshold2=200)

plt.figure(figsize=(12,6))

plt.subplot(131)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(132)
plt.title("Blurred Image")
plt.imshow(cv2.cvtColor(blur_image,cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(133)
plt.title("Canny Edge of Image")
plt.imshow(cv2.cvtColor(edges,cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.tight_layout()
plt.show()
```
