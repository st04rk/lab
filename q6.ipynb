```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def display_image_in_color_modes(image_path):
    img = cv2.imread(image_path)
    rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    grayscale_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    hsv_img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    ycrcb_img = cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
    lab_img = cv2.cvtColor(img,cv2.COLOR_BGR2LAB)

    plt.figure(figsize=(12,6))

    plt.subplot(231)
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
    plt.title("BGR (Original)")
    plt.axis("off")

    plt.subplot(232)
    plt.imshow(rgb_img)
    plt.title("RGB")
    plt.axis("off")

    plt.subplot(233)
    plt.imshow(grayscale_img,cmap='gray')
    plt.title("Grayscale")
    plt.axis("off")

    plt.subplot(234)
    plt.imshow(hsv_img)
    plt.title("HSV")
    plt.axis("off")

    plt.subplot(235)
    plt.imshow(ycrcb_img)
    plt.title("YCrCb")
    plt.axis("off")

    plt.subplot(236)
    plt.imshow(lab_img)
    plt.title("LAB")
    plt.axis("off")

    plt.show()

image_path="flora.jpeg"
display_image_in_color_modes(image_path)
```
